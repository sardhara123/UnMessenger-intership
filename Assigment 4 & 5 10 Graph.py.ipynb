```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\rahul.sardhara\Desktop\Total Three Month Report\Rahul Data\Data Science\Assignment 4 Dataset 1 (1205).csv')
```

```python
df.head()
```

```python
x=df['room_type']
y=df["price"]
```

```python
#1 Bar Plot
plt.bar(x,y,color="g",width=0.8)
plt.title('room type V/s price')
plt.show()
#Room Type Vs Rent Bar Chart#
```

```python
#2 Line Plot
x=df['New Year']
y=df["room_type"]
plt.figure(figsize=(15,4))
plt.plot(x,y,color="b")
plt.title('Year V/s room type')
plt.show()
#Year V/s room type used Line Chart#
```

```python
#3 Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=df)
plt.xlabel('neighbourhood_group')
plt.ylabel('price')
plt.title('Box Plot')
plt.show()
#Group V/s Price
```

```python
#4 Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['New Year'], bins=10)
plt.xlabel('New Year')
plt.ylabel('price')
plt.title('Histogram')
plt.show()
#Year V/s Price
```

```python
df2=pd.read_csv(r'C:\Users\rahul.sardhara\Desktop\Total Three Month Report\Rahul Data\Data Science\Assignment 4 Dataset 22 (1205).csv')
df2
```

```python
#5 Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df2['State'], df2['Salary'])
plt.xlabel('State')
plt.ylabel('Salary')
plt.title('Scatter Plot')
plt.show()
#State V/s Salary
```

```python
#6 Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Department', y='Salary', data=df2)
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Violin Plot')
plt.show()
# Department V/s Salary
```

```python
#7 Pie Chart
plt.figure(figsize=(10, 5))
plt.pie(df2['Sex'].value_counts(), labels=df2['Sex'].value_counts().index, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
# Male Female Comparison
```

```python
#8 SNS Chart
sns.displot(
    data=df2,
    x="Salary", y="PerformanceScore",hue="PerformanceScore")
#PerformanceScore Vs Salary Comparison
```

```python
#9 stripplot
sns.stripplot(y = df2['Salary'],x=df2['Sex'])
plt.show()
#Male Female Strip Plot agaist Salary 
```

```python
#10 Facegrid Plot
g = sns.FacetGrid(df2, col="MaritalDesc")
g.map(sns.histplot, "Sex")
plt.show()
#Comparison of Male Female woth Marital Status
```
